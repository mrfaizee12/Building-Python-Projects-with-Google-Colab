**Project 5: Hangman Game pro # 05**

```python
import random

# Hangman ASCII stages (visual representation)
hangman_stages = [
    """
       -----
       |   |
           |
           |
           |
           |
    =========
    """,
    """
       -----
       |   |
       O   |
           |
           |
           |
    =========
    """,
    """
       -----
       |   |
       O   |
       |   |
           |
           |
    =========
    """,
    """
       -----
       |   |
       O   |
      /|   |
           |
           |
    =========
    """,
    """
       -----
       |   |
       O   |
      /|\\  |
           |
           |
    =========
    """,
    """
       -----
       |   |
       O   |
      /|\\  |
      /    |
           |
    =========
    """,
    """
       -----
       |   |
       O   |
      /|\\  |
      / \\  |
           |
    =========
    GAME OVER!
    """
]

# Word categories
word_categories = {
    "fruits": ["apple", "banana", "grape", "orange", "mango"],
    "animals": ["tiger", "elephant", "giraffe", "panda", "zebra"],
    "countries": ["pakistan", "canada", "brazil", "germany", "france"]
}

def choose_word():
    print("\nğŸŒ Choose a category: Fruits, Animals, Countries")
    category = input("ğŸ‘‰ Enter category: ").strip().lower()

    if category in word_categories:
        return random.choice(word_categories[category])
    else:
        print("âŒ Invalid category! Defaulting to 'fruits'...")
        return random.choice(word_categories["fruits"])

def hangman():
    print("\nğŸ® Welcome to the Hangman Game!")
    secret_word = choose_word()
    guessed_letters = []
    attempts = 6  # Maximum attempts

    while attempts > 0:
        display_word = "".join([letter if letter in guessed_letters else "_" for letter in secret_word])
        print(f"\nğŸ”¤ Word: {display_word}")
        print(hangman_stages[6 - attempts])

        if "_" not in display_word:
            print("ğŸ‰ Congratulations! You guessed the word correctly!")
            return

        guess = input("\nâœï¸ Guess a letter: ").strip().lower()

        if len(guess) != 1 or not guess.isalpha():
            print("âŒ Invalid input! Enter a single letter.")
            continue

        if guess in guessed_letters:
            print("âš ï¸ You already guessed this letter! Try again.")
            continue

        guessed_letters.append(guess)

        if guess in secret_word:
            print("âœ… Good job! You found a letter!")
        else:
            print("âŒ Wrong guess! Try again.")
            attempts -= 1

    print(hangman_stages[6])
    print(f"\nğŸ’” Game Over! The correct word was: {secret_word}")

# Start the game
hangman()
```
